```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("study_data_100.csv")
df.head()
```

```python
df.shape, df.size
```

```python
X = df["studytime"]
Y = df["score"]
L = 0.001
```

```python
plt.plot(X, Y, '.')
plt.xlabel("Study Time")
plt.ylabel("Score")
plt.show()
```

```python
m = 0
c = 0
J = 0

# Calculate yi(pred) = h(xi) = mxi + c
# Calculate J = (1/2)sum((h(xi) - yi)^2)
# Update m := m - Lsum((h(xi)-yi)*xi)
# Update c := c - Lsum((h(xi)-yi))
# Repeat. Check if J is getting lower. If yes, then repeat until J < 0.1.

for i in range(df.shape[0]):
  J = J + 0.5*((m*X[i] + c - Y[i])**2)
print("Initial J = ", J)

J_old = J
J_new = 0
while J_new - J_old > 0.001:
  J_old = J_new
  for i in range(df.shape[0]):
    m = m - L * (m*X[i] + c - Y[i]) * X[i]
    c = c - L * (m*X[i] + c - Y[i])
  J_new = 0
  for i in range(df.shape[0]):
    J_new = J_new + 0.5*((m*X[i] + c - Y[i])**2)
  # Used the following to debug
  # print(J_new)
```

```python
plt.plot(X, m*X+c)
plt.plot(X, Y, '.')
plt.show()
```
